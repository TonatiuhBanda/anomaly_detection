```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Input
from keras import backend as K

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
```

```python
df_raw = pd.read_pickle("dataset_2d_points.pkl")
df_raw.head(3)
```

```python
df_raw.shape
```

```python
df_smart = df_raw.iloc[:1500, :2]
df_smart_labels = df_raw.iloc[:1500, 3]

df_validate = df_raw.iloc[1500:, :2]
df_validate_labels = df_raw.iloc[1500:, 3]

x_train, x_test, y_train, y_test = train_test_split(df_smart, df_smart_labels,
                                                    test_size=0.2, random_state=8)

x_val, y_val = df_validate, df_validate_labels

print(x_train.shape)
print(y_train.shape)
```

```python
model = Sequential()
model.add(Dense(6, input_dim=2, activation='tanh'))
model.add(Dense(1, activation='tanh'))
model.compile(loss='binary_crossentropy', optimizer='SGD', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=1500, batch_size=10)
```

```python
#y_pred = model.predict(x_test)
y_pred = model.predict(x_val)
```

```python
def get_class(threshold, y):
    if y >= threshold:
        return 1
    else:
        return 0

df_y_pred = pd.DataFrame(y_pred)
df_y_pred['pred_binary'] = df_y_pred.apply(lambda x: get_class(0, x[0]), axis=1)
df_y_pred.reset_index(drop=True, inplace=True)
#df_y_test = y_test.reset_index()
df_y_test = y_val.reset_index()
df_y_pred_test = pd.concat([df_y_pred, df_y_test], axis=1)
df_y_pred_test.head(3)
```

```python
auc = roc_auc_score(df_y_pred_test['pred_binary'], df_y_pred_test['class_2'])
print("AUC: {:.2%}".format(auc))
```

```python
tn, fp, fn, tp = confusion_matrix(df_y_pred_test['class_2'], df_y_pred_test['pred_binary']).ravel()
print("TN: {}, FP: {}, FN: {}, TP: {}".format(tn, fp, fn, tp))
```

```python
color_point = ['green', 'red']
color_test = df_raw['class_2'].apply(lambda x: color_point[x])
plt.scatter(df_raw['x'], df_raw['y'], marker='.',c=color_test )
plt.show()
```

```python
color_option = ['green', 'red', 'black', 'cyan']
def get_color(test, pred, color_option):
    if test == 1 and pred == 1:
        return color_option[1]
    elif test == 1 and pred == 0:
        return color_option[2]
    elif test == 0 and pred == 1:
        return color_option[3]
    else:
        return color_option[0]
df_y_pred_test['color'] = df_y_pred_test.apply(lambda row: get_color(row['class_2'], row['pred_binary'], color_option), axis=1)
df_y_pred_test.head(4)
```
